# Lab 5
Image filtering and sharpening using the parrot image.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Candidate paths for the parrot image (works from different working directories)
PARROT_CANDIDATES = [
    "parrot.jpg",
    "pdf_imgs/parrot.jpg",
    "lab 5/parrot.jpg",
    "lab 5/pdf_imgs/parrot.jpg",
]


def load_parrot(mode="color"):
    # Load parrot image in color or grayscale
    flag = cv2.IMREAD_COLOR if mode == "color" else cv2.IMREAD_GRAYSCALE

    for path in PARROT_CANDIDATES:
        image = cv2.imread(path, flag)
        if image is not None:
            # Convert BGR to RGB only for color display
            if mode == "color":
                return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            return image

    raise FileNotFoundError("Could not find parrot.jpg in Lab 5 paths.")


def show_images(images, titles, figsize=(15, 5), cmap=None):
    # Display images side by side
    plt.figure(figsize=figsize)
    for i, (img, title) in enumerate(zip(images, titles), 1):
        plt.subplot(1, len(images), i)
        if img.ndim == 2:
            plt.imshow(img, cmap=("gray" if cmap is None else cmap))
        else:
            plt.imshow(img)
        plt.title(title)
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

# Task 1: Unsharp Masking
Apply unsharp masking to the parrot image.


```python
# Load RGB parrot image
image = load_parrot(mode="color")

# Unsharp masking parameters
sigma = 2
amount = 1.5

# Normalize image to [0,1] for stable processing
image_float = image.astype(np.float32) / 255.0

# Blur image using Gaussian filter
blurred = cv2.GaussianBlur(image_float, (0, 0), sigmaX=sigma, sigmaY=sigma)

# Unsharp mask formula: original + amount*(original - blurred)
sharpened = np.clip(image_float + amount * (image_float - blurred), 0, 1)

# Show original, blurred, and sharpened
show_images(
    [image_float, blurred, sharpened],
    ["Original Image", "Blurred Image", "Sharpened (Unsharp Masking)"]
)
```

## Assessment Task 1
Convolve an image with a 7x7 box filter.


```python
# Load grayscale parrot image
img = load_parrot(mode="gray")

# Create normalized 7x7 box kernel
kernel = np.ones((7, 7), np.float32) / 49

# Apply box filter
smoothed_box = cv2.filter2D(img, -1, kernel)

# Show original and smoothed results
show_images(
    [img, smoothed_box],
    ["Original Image", "Smoothed Image (7x7 Box Filter)"],
    figsize=(12, 5),
    cmap="gray"
)
```

# Task 2: Gaussian Filter Comparison
Apply a 5x5 Gaussian filter and a 21x21 Gaussian filter.


```python
# Load grayscale parrot image
img2 = load_parrot(mode="gray")

# Apply Gaussian smoothing with two kernel sizes
gauss_5 = cv2.GaussianBlur(img2, (5, 5), 0)
gauss_21 = cv2.GaussianBlur(img2, (21, 21), 0)

# Show original and both Gaussian outputs
show_images(
    [img2, gauss_5, gauss_21],
    ["Original Image", "Gaussian 5x5", "Gaussian 21x21"],
    figsize=(15, 5),
    cmap="gray"
)
```

# Task 3: Laplacian Sharpening
Apply Laplacian sharpening and show original, Laplacian result, and sharpened image.


```python
# Load grayscale parrot image
img3 = load_parrot(mode="gray")

# Convert to float [0,1] before Laplacian
image_float = img3.astype(np.float32) / 255.0

# Compute Laplacian response
laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)

# Sharpen by subtracting Laplacian
sharpened_laplacian = np.clip(image_float - laplacian, 0, 1)

# Show original, Laplacian, and sharpened
show_images(
    [image_float, laplacian, sharpened_laplacian],
    ["Original Image", "Laplacian (3x3)", "Sharpened Image"],
    cmap="gray"
)
```
